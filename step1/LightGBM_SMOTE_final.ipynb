```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
import os
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, f1_score
from imblearn.over_sampling import SMOTE
from lightgbm import LGBMClassifier

# ===== 0. ËÆÄÂÖ•Ë≥áÊñô‰∏¶Âêà‰Ωµ serial_no =====
serials_df = pd.read_excel("../data/Ê•≠ÂãôÂì°Ë≥áÊñô0325V01.xlsx", sheet_name="DATA", engine="openpyxl", usecols=[0])
serials_df.columns = ["serial_no"]
file_path = "../data/filled_data_label_numeric.xlsx"
filled_data_label_numeric_df = pd.read_excel(file_path, engine='openpyxl')
filled_data_label_numeric_df = pd.concat([serials_df, filled_data_label_numeric_df], axis=1)

# ===== 1. ÁâπÂæµËàáÁõÆÊ®ô =====
X = filled_data_label_numeric_df.drop(columns=['serial_no', 'abnormal_target'])
y = filled_data_label_numeric_df['abnormal_target']

# ===== 2. Ê®ôÊ∫ñÂåñ =====
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ===== 3. GridSearch Áî®ÂèÉÊï∏ÁØÑÂúç =====
param_grid = {
    'n_estimators': [100, 200],
    'learning_rate': [0.05, 0.1],
    'max_depth': [3, 5],
    'num_leaves': [15, 31]
}

# ===== 4. ÊêúÂ∞ãÊúÄ‰Ω≥Ê®°ÂûãÔºàÈåØË™§Êï∏ÊúÄÂ∞ëÔºâ=====
best_model = None
best_seed = None
best_params = None
best_misclassified_count = float('inf')
best_misclassified_df = None
results = []

for seed in range(15):
    print(f"\nüå± Seed {seed}:")
    
    X_train, X_test, y_train, y_test = train_test_split(
        X_scaled, y, test_size=0.2, random_state=seed, stratify=y)
    
    smote = SMOTE(random_state=seed)
    X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)
    X_train_resampled_df = pd.DataFrame(X_train_resampled, columns=X.columns)

    grid = GridSearchCV(
        estimator=LGBMClassifier(random_state=seed, verbose=-1),
        param_grid=param_grid,
        scoring='f1_macro',
        cv=3,
        n_jobs=-1
    )
    grid.fit(X_train_resampled_df, y_train_resampled)
    best_grid_model = grid.best_estimator_

    y_pred = best_grid_model.predict(X_test)
    f1 = f1_score(y_test, y_pred, average='macro')

    # ‰øÆÊ≠£ÈåØË™§ÔºöÈáçÊñ∞Â∞çÈΩäÁ¥¢Âºï
    y_test_reset = y_test.reset_index(drop=True)
    y_pred_series = pd.Series(y_pred)
    mismatch_index = y_test_reset != y_pred_series

    wrong_rows = filled_data_label_numeric_df.iloc[y_test.index[mismatch_index.to_numpy()]].copy()
    wrong_rows["true_label"] = y_test_reset[mismatch_index].values
    wrong_rows["predicted_label"] = y_pred_series[mismatch_index].values
    wrong_rows["seed"] = seed
    wrong_rows["t0p1"] = ((wrong_rows["true_label"] == 0) & (wrong_rows["predicted_label"] == 1)).astype(int)
    wrong_rows["t1p0"] = ((wrong_rows["true_label"] == 1) & (wrong_rows["predicted_label"] == 0)).astype(int)

    mis_count = len(wrong_rows)
    print(f"‚ùå ÈåØË™§Á≠ÜÊï∏Ôºö{mis_count}")
    print("ÈåØË™§ serial_noÔºö", wrong_rows["serial_no"].tolist())

    results.append({"seed": seed, "f1_macro": f1, "misclassified": mis_count})

    if mis_count < best_misclassified_count:
        best_model = best_grid_model
        best_seed = seed
        best_params = grid.best_params_
        best_misclassified_count = mis_count
        best_misclassified_df = wrong_rows.copy()

# ===== 5.1 F1 ÂàÜÊï∏ËàáÈåØË™§Êï∏Áµ±Ë®àÂúñ =====
results_df = pd.DataFrame(results)
plt.figure(figsize=(8, 4))
plt.plot(results_df['seed'], results_df['f1_macro'], marker='o', label='F1 Score')
plt.plot(results_df['seed'], results_df['misclassified'], marker='x', color='red', label='Misclassified Count')
plt.title('Seed vs F1 & Error Count')
plt.xlabel('Random Seed')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# ===== 5.2 Ê∑∑Ê∑ÜÁü©Èô£ÂèØË¶ñÂåñÔºà3 x 5Ôºâ=====
fig, axes = plt.subplots(5, 3, figsize=(15, 20))
axes = axes.flatten()

for i, row in results_df.iterrows():
    seed = int(row['seed'])
    
    X_train, X_test, y_train, y_test = train_test_split(
        X_scaled, y, test_size=0.2, random_state=seed, stratify=y)

    smote = SMOTE(random_state=seed)
    X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)
    X_train_resampled_df = pd.DataFrame(X_train_resampled, columns=X.columns)

    model = LGBMClassifier(random_state=seed, verbose=-1, **(best_params if seed == best_seed else {}))
    model.fit(X_train_resampled_df, y_train_resampled)
    y_pred = model.predict(X_test)

    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False,
                xticklabels=["Normal (0)", "Abnormal (1)"],
                yticklabels=["Normal (0)", "Abnormal (1)"],
                ax=axes[i])
    axes[i].set_title(f"Seed {seed}")
    axes[i].set_xlabel("Predicted")
    axes[i].set_ylabel("Actual")

plt.tight_layout()
plt.show()

# ===== 6. ÂÑ≤Â≠òÊúÄ‰Ω≥Ê®°ÂûãËàá scaler =====
os.makedirs("../model", exist_ok=True)
joblib.dump(best_model, "../model/best_lgbm_model.pkl")
joblib.dump(scaler, "../model/final_scaler.pkl")
print(f"\n‚úÖ Â∑≤ÂÑ≤Â≠òÊúÄ‰Ω≥Ê®°ÂûãÔºàseed={best_seed}, ÈåØË™§Êï∏={best_misclassified_count}Ôºâ")

# ===== 7. Ëº∏Âá∫ÊúÄ‰Ω≥Ê®°ÂûãÈ†êÊ∏¨ÈåØË™§ÁöÑ row =====
best_misclassified_df.to_csv("../data/best_model_misclassified_rows.csv", index=False)
print(f"‚úÖ ÊúÄ‰Ω≥Ê®°ÂûãÈåØË™§Ê®£Êú¨Â∑≤Ëº∏Âá∫ÔºåÂÖ± {len(best_misclassified_df)} Á≠Ü")

```

```python
import joblib
import numpy as np

# === ÊñπÊ≥ïÔºöËºâÂÖ•Ê®°Âûã‰∏¶È†êÊ∏¨ÊüêÊ•≠ÂãôÂì°ÁöÑÁï∞Â∏∏Ê©üÁéá ===
def predict_abnormal_probability(serial_no, df):
    # ËºâÂÖ•Ê®°ÂûãËàá scaler
    model = joblib.load("../model/best_lgbm_model.pkl")
    scaler = joblib.load("../model/final_scaler.pkl")

    # Êì∑ÂèñÊåáÂÆö serial_no ÁöÑ row
    target_row = df[df["serial_no"] == serial_no]
    if target_row.empty:
        print(f"‚ùó Êâæ‰∏çÂà∞ serial_no = {serial_no}")
        return None

    # ÂéªÊéâ serial_no Ëàá abnormal_target Ê¨Ñ‰Ωç
    features = target_row.drop(columns=["serial_no", "abnormal_target"])
    
    # Ê®ôÊ∫ñÂåñ
    features_scaled = scaler.transform(features)

    # È†êÊ∏¨Ê©üÁéáÔºà[0]=normal, [1]=abnormalÔºâ
    prob = model.predict_proba(features_scaled)[0][1]
    print(f"üîç {serial_no} ÁÇ∫ abnormal_target=1 ÁöÑÊ©üÁéáÁÇ∫Ôºö{prob:.4f}")
    return prob

# === ÁØÑ‰æãÂü∑Ë°å ===
predict_abnormal_probability("agnt_3000", filled_data_label_numeric_df)
```
